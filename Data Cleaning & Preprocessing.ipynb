# 📌Step 3 — Handle Missing Values (Very Important)

```python
df.isnull().sum()
```

# Age Column

```python
df['age'].fillna(df['age'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

# Year of Experience Column

```python
df['years_of_experience'].fillna(df['years_of_experience'].median(), inplace=True)
```

```python
df.isnull().sum()
```

# hourly_rate (USD)

```python
df['hourly_rate (USD)'].fillna(df['hourly_rate (USD)'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

# rating 

```python
df['rating'].fillna(df['rating'].mean().round(1), inplace=True)
```

```python
df.isnull().sum()
```

# is_active

```python
df['is_active'].fillna(df['is_active'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

# client_satisfaction

```python
df['client_satisfaction'].fillna(df['client_satisfaction'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

# 📌Step 4 — Fix Data Types

```python
df['age'] = df['age'].astype(int)
```

```python
df['years_of_experience'] = df['years_of_experience'].astype(int)
```

```python
df.info()
```

# 📌 Step 5 — Fix Wrong or Inconsistent Values

```python
df.head(4)
```

# Before cleaning

```python
df['gender'].value_counts()
```

```python
df['gender'] = df['gender'].replace({
    'FEMALE':'Female',
    'f':'Female',
    'F':'Female',
    'female':'Female',

    'MALE':'Male',
    'M':'Male',
    'm':'Male',
    'male':'Male'    
})
```

# After cleaning

```python
df['gender'].value_counts()
```

# Before cleaning

```python
df['hourly_rate (USD)'].value_counts()
```

```python
df['hourly_rate (USD)'] = df['hourly_rate (USD)'].str.replace('$', '').str.replace('USD', '').astype(int)
```

# After cleaning

```python
df['hourly_rate (USD)'].value_counts()
```

# Before cleaning

```python
df['is_active'].value_counts()
```

```python
df['is_active'] = df['is_active'].replace({
    'Y':'1',
    'yes':'1',
    'True':'1',

    'N':'0',
    'no':'0',
    'False':'0'
    
}).astype(int)
```

# After cleaning

```python
df['is_active'].value_counts()
```

# client_satisfaction

```python
df['client_satisfaction'] = df['client_satisfaction'].str.replace('%','').astype(int)
```

```python
df.rename(columns={'client_satisfaction':'client_satisfaction(%)'}, inplace=True)
df.head(1)
```

# IQR method

```python
q1 = df['years_of_experience'].quantile(0.25)
q3 = df['years_of_experience'].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_r = q1 - (1.5*IQR)
max_r = q3 + (1.5*IQR)

min_r, max_r
```

```python
df.shape
```

```python
df = df[df['years_of_experience'] <= max_r]
```
